```python
from matplotlib import pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
plt.plot(years,gdp,color='blue',marker='o',linestyle='dotted')
plt.title('Nominal GDP')
plt.ylabel('Billions of Dollars')
plt.show()
```

```python
#Bar Chart
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"] 
num_oscars = [5, 11, 3, 8, 10]
plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")
plt.bar(movies,num_oscars,width=0.7,color='black')
plt.show()
```

```python
#Bar Chart
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"] 
num_oscars = [5, 11, 3, 8, 10]
xs=[x for x,_ in enumerate(movies)]
print(xs)
plt.bar(xs,num_oscars,width=0.7,color='purple')
plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")
plt.xticks([i for i ,_ in enumerate(movies)],movies)
plt.show()
```

```python
#Bar Chart 2
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
function= lambda x: (x//10)*10
histogram= Counter(function(grade) for grade in grades)
plt.bar(histogram.keys(),histogram.values(),width=4)
plt.xticks([i*10 for i in range (11)])
plt.axis([-5, 105, 0, 5]) 
plt.xlabel("Decile") 
plt.ylabel("# of Students") 
plt.title("Distribution of Exam 1 Grades") 
plt.show()
```

```python
import numpy as np
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0,45,55,35,16,11,24,43,77,3,9,17,36]
function=lambda x: (x//10)*10
values=np.array([function(grade) for grade in grades])
fail=len(np.take(values,np.where(values<=50)))
sixty_s=len(np.take(values,np.where(values==60)))
seventy_s=len(np.take(values,np.where(values==70)))
a_grade=len(np.take(values,np.where(values==80)))
a_plus=len(np.take(values,np.where(values==90)))

x_axis=["FAIL","C","B","A","A+"]
y_axis=[fail,sixty_s,seventy_s,a_grade,a_plus]

plt.bar(y,[y for y,_ in enumerate(y_axis)])
plt.show
```

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends,minutes,color='purple')
plt.show

#labeling the points
for friend,minute,label in zip(friends,minutes,labels):
    plt.annotate( label,
                 xy=(friend,minute),
                xytext=(5,-5),
                textcoords='offset points' 
                )
plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends") 
plt.ylabel("daily minutes spent on the site") 
plt.show()
```
