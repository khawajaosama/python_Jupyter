```python
import datetime
def dates2num_2(s):
    return datetime.datetime.strptime(str(s),"%d-%m-%Y").date().weekday()
```

```python
dates2num_2('28-01-2011')
```

```python
import numpy as np
```

```python
dates, price, close=np.loadtxt('prices.csv', delimiter=',', usecols=(0,2,3), converters={1: dates2num_2}, unpack=True, dtype='str')
dates=list(map(dates2num_2, dates))
dates=np.asanyarray(dates,'float64')
dates
```

```python
price=np.asanyarray(price,'f')
price
```

```python
zeros=np.zeros(5,dtype='float')
zeros
```

```python
close=np.asanyarray(price,'d')
close
```

```python
for i in range(5):
    indices=np.where(dates==i)
    prices=np.take(close,indices)
    avg=np.mean(prices)
    days=['Monday','Tuesday','Wednesday','Thursday','Friday']
    print('Day=',days[i],' Price=',prices,' Average=',avg)
    zeros[i]=avg
```

```python
zeros
```

```python
top = np.max(zeros)
print("Highest average", top)
print("Top day of the week", np.argmax(zeros))
bottom = np.min(zeros)
print("Lowest average", bottom)
print("Bottom day of the week", np.argmin(zeros))
```

```python
import requests
import datetime as dt
import matplotlib.pyplot as plt
def crypto_crncy_price(symbol, comparison_symbol = 'USD', all_data=True, aggregate=1,limit = 1, exchange=''):
    a = int(input('1 for Minute\n2 for Hourly\n3 for Today '))
    url = 'https://min-api.cryptocompare.com/data/histominute?fsym={}&tsym={}&limit={}&aggregate={}'.format(symbol.upper(), comparison_symbol.upper(), limit, aggregate)
    if a == 1: url = url.replace('&limit=1', '&limit=9999')
    elif a == 2: url = url.replace('histominute','histohour'); url = url.replace('&limit=1', '&limit=9999')
    elif a == 3: url = url.replace('histominute','histoday')
    if all_data and a == 3: url += '&allData=true'
    if exchange: url += '&e={}'.format(exchange)
    a = requests.get(url).json()['Data']
    b = [(a[i]['time'], a[i]['close'], dt.datetime.fromtimestamp(a[i]['time'])) for i in range(len(a))]
    plt.plot([i[2] for i in b], [i[1] for i in b])
    plt.xticks(rotation = 30)
    plt.show()

# function testing: 
a = print(crypto_crncy_price('btc'))
b = print(crypto_crncy_price('eth'))
c = print(crypto_crncy_price('neo'))
d = print(crypto_crncy_price('neo', 'pkr'))
for i in a,b,c,d: print(i)
```
